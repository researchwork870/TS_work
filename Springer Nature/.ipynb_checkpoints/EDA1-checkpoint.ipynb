```python
import pandas as pd
df = pd.read_csv("SPRBI-5213 - Order Data 002 2nd Half 2021 AT Processed.csv")
```

```python
def resumetable(df): 
    print(f"Dataset Shape: {df.shape}")
    summary = pd.DataFrame(df.dtypes, columns=['dtypes'])
    summary = summary.reset_index()
    summary['Name'] = summary['index']
    summary = summary[['Name', 'dtypes']]
    summary['Missing'] = df.isnull().sum().values
    summary['Missing(% of Total values)'] = (
        100 * summary['Missing']) / df.shape[0]
    summary['Uniques'] = df.nunique().values
    summary['First Value'] = df.loc[0].values
    summary['Second Value'] = df.loc[1].values
    return summary

summary = resumetable(df)
```

```python
import plotly.express as px
from plotly.offline import plot
from plotly.subplots import make_subplots
import plotly.graph_objects as go

def visualization_ts(product_id,frequency):
    ts["Invoice Date"] = pd.to_datetime(ts["Invoice Date"])
    ts.index = ts["Invoice Date"]
    ts_data = ts[ts['Product ID'] == product_id].resample(frequency)
    new = pd.DataFrame(ts_data["Total Sales Units consolidated"].sum()).reset_index()
    data_points = new.shape[0]
    fig = px.line(new
                 , x="Invoice Date"
                 , y="Total Sales Units consolidated"
                 , title = "TS(Product_id: {}, Observations: {})".format(product_id,data_points) 
                 , hover_data=['Invoice Date','Total Sales Units consolidated'])
    return fig.show()
```

```python
for i in list(ts_count[ts_count[0]>50]["Product ID"]):
    visualization_ts(i,"30D")
```

```python
for i in list(ts_count[ts_count[0]>50]["Product ID"]):
    visualization_ts(i,"7D")
```

```python
def monthly_pattern_over_years(df,time_column, time_dependent_variable, rotation=70, color="#D39200"):
    for year in df[time_column].dt.year.unique():
        # Filtering data for specific year
        x = df[df[time_column].dt.year == year]

        # Extracting month name from date
        x['month'] = x[time_column].dt.month_name()

        # Grouping data by months and adding time_dependent_variable
        x = x.groupby(['month'], as_index=False).agg(
            {time_dependent_variable: 'sum'})

        # Plotting month Vs time_dependent_variable for every year
        # plt.figure(figsize=(12,8))
        sns.barplot(x=x.month, y=time_dependent_variable, data=x, order=x.sort_values(
            time_dependent_variable).month, color=color)
        plt.xticks(rotation=rotation)
        plt.title(time_dependent_variable + " " +
                  "per Month ("+str(year)+")", fontsize=14)
        plt.ylabel(time_dependent_variable, fontsize=14)
        plt.xlabel('Month', fontsize=14)
        plt.show()
        del x


monthly_pattern_over_years(ts,"Invoice Date", "Total Sales Units consolidated")
```
